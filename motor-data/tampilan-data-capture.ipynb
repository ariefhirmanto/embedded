```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as scipy
from scipy import optimize
from matplotlib.ticker import AutoMinorLocator
from matplotlib import gridspec
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
# numerical data file
# sampling rate: 10 ms
# capture 1 sampai 3: gerakan cepat dan step-step
filename0="capture.txt" #tanpa kapasitor
filename1="capture1.txt" #tanpa kapasitor
filename2="capture2.txt" #kapasitor 100 nF
filename3="capture3.txt" #kapasitor 10uF
filename4="capture4.txt" # tegangan input 12 volt, ada masalah dengan serial output
filename5="capture5.txt" 
filename6="capture6.txt" 
# capture 4 sampai 6: gerakan lambat dan cepat


```

```python
# 8 volt
fig = plt.figure(figsize=(16,4))
gs = gridspec.GridSpec(1,1)
ax1 = fig.add_subplot(gs[0])
ax1.plot(data1, "r-")
ax1.plot(data2, "g-")
ax1.plot(data3, "b-")
fig.savefig("gabungan-8-volt.png", format="png",dpi=60)
```
